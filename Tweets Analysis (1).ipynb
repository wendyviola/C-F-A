# Data collection
Research for twitter hashtag #RacistEU for the period 26th Feb to March 26th 2022.
This was done through Python’s twint library tool
Data is stored in a CSV file
Some of the challanges faced is that some tweets were not being pulled in or rather one time the tool could pull all tweets then when I run the script again the number deplets

Following cell is a sample code used to collect data


```python
import twint
import nest_asyncio

nest_asyncio.apply()
# Configure
c = twint.Config()
c.Search = "#RacistEU"
#c.Retweets = True
c.Since = '2022-02-26'
c.Until = '2022-03-26'
c.Count = True #total number of hashtags fetched
c.Store_csv = True
c.Debug = True
c.Output = "Twitter Data"
# Run
twint.run.Search(c)
```

```python
#Developing the Wordcloud

import nltk # Natural Language ToolKit
nltk.download('stopwords')
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from wordcloud import WordCloud 


def plot_cloud(wordcloud):
    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud) 
    plt.axis("off");
wordcloud = WordCloud(width = 500, height = 500, background_color='pink', random_state=10).generate(final_result)
plot_cloud(wordcloud)
```
